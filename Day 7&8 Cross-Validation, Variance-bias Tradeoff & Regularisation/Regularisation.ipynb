```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
df = pd.read_csv('test.csv', names = ['X1','Y'])
```

```python
df
```

```python
model = LinearRegression()
model.fit(df[['X1']],df['Y'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model = LinearRegression()
model.fit(df[['X1']],df['Y'])

print("Slope: ",model.coef_)
print("Intercept: ",model.intercept_)

pred_df = model.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'B')
```

```python
#Lambda - Alpha
model_ = Ridge(alpha = 0.05)
model_.fit(df[['X1']],df['Y'])

print("Slope: ",model_.coef_)
print("Intercept: ",model_.intercept_)

pred_df = model.predict(df[['X1']])
pred_df_ = model_.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'B')
plt.plot(df.X1,pred_df_,color = 'green')
```

```python
#Lambda - Alpha
model_ = Lasso(alpha = 0.05)
model_.fit(df[['X1']],df['Y'])

print("Slope: ",model_.coef_)
print("Intercept: ",model_.intercept_)

pred_df = model.predict(df[['X1']])
pred_df_ = model_.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'B')
plt.plot(df.X1,pred_df_,color = 'green')
```

```python
#tuning to find the optimum value of lambda-alpha
for i in np.arange(0,0.05,0.001):
    print("i-value",i)
    model_ = Lasso(alpha = i)
    model_.fit(df[['X1']],df['Y'])
    
    #scores = cross_val_score(model_,X_train,train_y,scoring ='neg_mean_squared_error',cv=folds)
    #score_RMSE = [np.sqrt(i) for i in abs(scores)]
    
    print("Slope: ",model_.coef_)
    print("Intercept: ",model_.intercept_)

    pred_df = model.predict(df[['X1']])
    pred_df_ = model_.predict(df[['X1']])
    plt.scatter(df.X1,df.Y,color = 'red')
    plt.plot(df.X1,pred_df,'B')
    plt.plot(df.X1,pred_df_,color = 'green')
```
