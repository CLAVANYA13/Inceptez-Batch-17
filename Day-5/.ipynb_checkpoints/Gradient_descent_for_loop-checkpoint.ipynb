```python
#Try till 9:30
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

data = pd.read_csv("data.csv")
X = data.iloc[:,0]
Y = data.iloc[:,1]
plt.scatter(X,Y)
plt.show()
```

```python
#Build the Gradient Descent Model: 
#Step 1: initializing the values for m and c 
m = 0 
c = 0 

#Step 2: Initialize the learning RAte and iterations 
L = 0.0001
iter_ = 1000

n = float(len(X))
print(n)
#For loop - Performing Gradient Descent 
for i in range(iter_):
    Y_pred = (m*X)+c #The current prediction value 
    D_m = (-2/n)*sum(X * (Y - Y_pred))
    D_c = (-2/n)*sum(Y - Y_pred)
    m = m - (L*D_m) 
    c = c - (L*D_c)
    print("The Slope and intercept for iteration:",i+1)
    print("Slope: ",m)
    print("Intercept: ",c)
    
print("After 1000 iterations values of slope: %0.2f and intercept: %0.2f " %(m,c))
```

```python
#Making Predictions

Y_pred = m*X + c

plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color = 'red')
plt.show()
```
