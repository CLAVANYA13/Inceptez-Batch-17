```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = df.drop(columns = "Price")
```

```python
vif_df = pd.DataFrame()
vif_df['features'] = X.columns 

vif_df['VIF'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

```python
vif_df
```

```python
#Practically speaking around 20is ok
#Multicolinearity does not affect the fit directly 
```

```python
X = df.drop(columns = ["Price",'Avg. Area Number of Rooms'])
```

```python
vif_df = pd.DataFrame()
vif_df['features'] = X.columns 

vif_df['VIF'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

```python
vif_df
```

```python
y = df['Price']
```

## 1. Model defenition

```python
model = LinearRegression()
```

## 2. Model Fit - Model Train

```python
model.fit(X,y)
```

```python
model.coef_
#y = c+m1x1+m2x2+m3x3+m4x4+error
```

```python
model.intercept_
```

## 3. Model Prediction

```python
y_pred = model.predict(X)
```

```python
y_pred
```

## 4. Model Evaluation

```python
#SSE and MSE - Reduced 
print("MSE: ",mean_squared_error(y,y_pred))
```

```python
#Root Mean Squared Error 
print("RMSE: ",np.sqrt(mean_squared_error(y,y_pred)))
```

## Assumptions 
##### 1. Linear relationship between X & Y (correlation Graph)
##### 2. Little to no Multi-collinearity
##### 3. The residual variance should be of Homoscadacity (constant variance)
##### 4. Residuals are normally distributed
##### 5. The relationship between actual and predicted will also be linear 

```python
#5. Linearity between actual and predicted 
plt.xlabel("Actual house price")
plt.ylabel("Predicted house price")

plt.scatter(y,y_pred)
```

```python
#4. Normal distribution 
Residuals = y-y_pred 
sns.distplot(Residuals)
```

```python
#3 homoscadacity 
plt.scatter(y_pred,Residuals)
```
