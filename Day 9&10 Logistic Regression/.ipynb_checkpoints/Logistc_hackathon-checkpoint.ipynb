```python
#Filling the NA values 
1. Column - Married. Fill Null values with mode of Married column
2. If dependents can be filled as the subset of mode of group by married
3. Credit history if loan_Status is 1 then credit history is - 1 else - 0 
    - test - Mode 
4. Fill Null values with mode of Gender column
```

```python
#f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else x.mode().iloc[0]
#df = df.fillna(df.groupby('G').transform(f))
```

```python
#Married 
data['Married'].fillna(data['Married'].value_counts().index[0], inplace=True)
```

```python
#Dependents
data['Dependents'] = data.groupby('Married').Dependents.transform(lambda x: x.fillna(x.mode()[0]))
```

```python
data['Credit_History'].value_counts()
```

```python
#Credit history Train 
data.loc[(data['source']=='train') & (data['Credit_History'].isna()==True) & (data['Loan_Status']=='Y'),'Credit_History']=1.0
data.loc[(data['source']=='train') & (data['Credit_History'].isna()==True) & (data['Loan_Status']=='N'),'Credit_History']=0.0
```

```python
data.isna().sum().sort_values(ascending = False)
```

```python
#Credit history Test 
data['Credit_History'].fillna(data['Credit_History'].value_counts().index[0], inplace=True)
```

```python
#Gender with mode 
data['Gender'].fillna(data['Gender'].value_counts().index[0], inplace=True)
```

```python
#Self_employed with mode 
data['Self_Employed'].fillna(data['Self_Employed'].value_counts().index[0], inplace=True)
```

```python
data.groupby(['Gender','Married','Education','Dependents'])['Loan_Amount_Term'].median()
```

```python
data['Loan_Amount_Term'] = data.groupby(['Gender','Married','Education','Dependents']).Loan_Amount_Term.transform(lambda x: x.fillna(x.median()))
```

```python
# Variable - LoanAmount - Identify on which other columns LoanAmount depends
# Explore LoanAmount vs categorical variables
g = sns.factorplot(y="LoanAmount",x="Credit_History",data=data,kind="box",hue="Gender")
g = sns.factorplot(y="LoanAmount",x="Property_Area",data=data,kind="box", hue="Dependents")
g = sns.factorplot(y="LoanAmount",x="Gender", data=data,kind="box",hue="Married")
g = sns.factorplot(y="LoanAmount",x="Married", data=data,kind="box")
g = sns.factorplot(y="LoanAmount",x="Education", data=data,kind="box")
g = sns.factorplot(y="LoanAmount",x="Self_Employed", data=data,kind="box")
g = sns.factorplot(y="LoanAmount",x="Dependents", data=data,kind="box")
```

```python
data['LoanAmount'] = data.groupby(['Property_Area','Gender','Married','Education','Self_Employed','Dependents']).LoanAmount.transform(lambda x: x.fillna(x.median()))
```

```python
data.isna().sum().sort_values(ascending = False)
```

```python
data["Dependents"] = data["Dependents"].map({"0": 0, "1": 1,"2": 2, "3+": 3})

data["Property_Area"] = data["Property_Area"].map({"Rural":0, "Semiurban":1, "Urban": 2,})
```

```python
data['LoanAmount'] = data['LoanAmount'] * 1000
```
